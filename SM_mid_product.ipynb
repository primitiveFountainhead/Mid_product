Each node contains the value generated xn which is generated by multiplying x(n-1) and x(n-2), the two previous nodes. The value in the edge represnts the product x(n-1) * x(n-2).

```python
import logging

logging.basicConfig(format='[%(levelname)s@%(name)s] %(message)s', level=logging.DEBUG)
```

```python
from graphviz import Digraph

# searchNotAdj(list, x0, x1) checks if x0 and x1 comes within list in adjescently, indicating a loop
def searchNotAdj(list, x0, x1):
    for i in range(len(list) - 1):
        if list[i] == x0 and list[i + 1] == x1:
            return False
    return True

#extract digits in N between ms and lsth digit, inclusive. ms is assumed to be greater than or equal to ls
def extractDigit(N, ms, ls):
    dms = 10**(ms) 
    dls = 10**(ls - 1)
    return int((N%dms)/dls)

# midPrdct(seed1, seed2, k, count) generates the id product sequence using seed1 and seed2 as start, 
#extracting k digits for a max count of count.
#returns a tuple V,E V is vertex and E is Edges, whic are the product.
def midPrdct(seed1, seed2, k, count):
    V = []
    E= []
    x0 = seed1
    x1 = seed2
    prdct =x0*x1
    

    MS = 3*k/2
    LS = k/2 + 1

    V.append(x0)
    V.append(x1)
    E.append(0)
    x0 = x1
    x1 = extractDigit(prdct, MS, LS)
    cnt = 0
    while searchNotAdj(V, x0, x1) and cnt < count  :
        V.append(x1)
        E.append(prdct)
        prdct = x0*x1
        x0 = x1
        x1 = extractDigit(prdct, MS, LS)
        cnt = cnt + 1

    V.append(x1)
    E.append(prdct)
      
    

    return V, E

#Class used to draw the grap
class drawGraph:
    dot = Digraph(comment='midSqr', strict = True, format = 'svg')

    # Draw a single graph, take input V, E, and maps V[i] to V[i + 1] via edge E[i]
    def add(self, V, E):
        self.dot.node(str(V[0]) + '|' + str(V[1]), label= 'seed1='+ str(V[0]) + '\nseed2='+ str(V[1]), style='filled', color = 'green') 
        for i in range(2,len(V)):
            self.dot.edge(str(V[i-2]) + '|' + str(V[i-1]),str(V[i-1]) + '|' + str(V[i]), label = str(E[i-1]))
            self.dot.node(str(V[i-1]) + '|' + str(V[i]), label=  str(V[i]) )
        self.dot.node(str(V[len(V)-2]) + '|' + str(V[len(V)-1]), label=  str(V[len(V)-1]), style = 'filled', color = 'red' )
        
    

        
#main

k = int(input('Enter the number of digits to be extracted: '))
seed1 = int(input('Enter the first number to be used as seed: '))
seed2 = int(input('Enter the second number to be used as seed: '))
count = int(input('Enter the maximum number of iteration to go through: '))

V,E = midPrdct(seed1,seed2, k,count)

D = drawGraph()
D.add(V,E)
D.dot



```
